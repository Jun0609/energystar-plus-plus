```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
import os
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

from html_table_extractor.extractor import Extractor
from bs4 import BeautifulSoup

import requests
from os import listdir
from os.path import isfile, join


%matplotlib inline
```

```python
def extract_table_2cols(html, tabId):
    soup = BeautifulSoup(html, 'lxml')
    table = soup.find('table', attrs={'id':tabId})
    if(table is None):
        return None
    
    atrs = {}
    for row in table.find_all('tr'):
        cols = row.find_all('td')
        c1 = cols[0].text.strip()
        c2 = cols[1].text.strip()
        #print type(cols[0]), len(cols)        
        atrs[c1] = c2
        #print [item.text.strip() for item in row.find_all('td')]    
    return atrs
```

```python
save_dir = "../data/seatle/tax/"
onlyfiles = [f for f in os.listdir(save_dir) if os.path.isfile(os.path.join(save_dir, f))]
count = len(onlyfiles)

print "%d files " % (count)

atrs_list = []

c = 0
for filepath in onlyfiles:
    c = c + 1
    print '%5d / %5d - %s' % (c, count, filepath), 
    
    html = ''
    with open(os.path.join(save_dir, filepath), 'r') as f:
        html = f.read()    
    atrs = extract_table_2cols(html, 'kingcounty_gov_cphContent_DetailsViewCommBldg')
    
    if atrs is not None:
        print len(atrs)
        atrs['parcelId'] = filepath.replace(".html", "")
        atrs_list.append(atrs)
    else:
        print 'No data found!'
    #break
```

```python
df = pd.DataFrame(atrs_list)
df.to_csv("../data/seatle/property_assessment_metadata_raw.csv", index = True, float_format='%.02f')
print df.shape
df.head()
```
