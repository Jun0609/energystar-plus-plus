```python
# handling ASP's page numbers
# https://scraperwiki.com/2011/11/how-to-get-along-with-an-asp-webpage/
    
save_dir = "../data/seatle/tax/"
def download_and_save(parcelId, url):
    try:
        r = requests.get(url)
        print parcelId, r.status_code
        filename = os.path.join(save_dir, parcelId + '.html')
        with open(filename, 'wb') as f:
            f.write(r.content)
    except Exception as e:
        print '%s Exception %s' % (parcelId, str(e))
```

```python
#http://blue.kingcounty.com/Assessor/eRealProperty/Detail.aspx?ParcelNbr=2125049015
url0 = 'http://blue.kingcounty.com/Assessor/eRealProperty/Detail.aspx?ParcelNbr='
for ix, row in edata.iterrows():
    if ix < 2088:
        continue
    parcelId = row['TaxParcelIdentificationNumber']
    url1 = url0 + parcelId
    #print ix, parcelId
    download_and_save(parcelId,url1)
```
